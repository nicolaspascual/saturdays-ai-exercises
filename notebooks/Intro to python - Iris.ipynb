```python
with open('./data/iris.data', 'r') as f:
    lines = f.readlines()
lines
```

```python
cleaned_lines = [line.strip() for line in lines]
cleaned_lines
```

```python
data = [line.split(',') for line in cleaned_lines if line != '']
data
```

```python
class IrisFlower(object):
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width, iris_type):
        self.sepal_length = sepal_length
        self.sepal_width = sepal_width
        self.petal_length = petal_length
        self.petal_width = petal_width
        self.iris_type = iris_type
    
    def get_petal(self):
        return self.petal_length, self.petal_width

    def get_sepal(self):
        return self.sepal_length, self.sepal_width

    def __repr__(self):
        return f'{self.iris_type} Sepal({self.sepal_length}, {self.sepal_width})'\
            + f' Petal({self.petal_length}, {self.petal_width})'
```

```python
flowers = [
    IrisFlower(float(row[0]), float(row[1]), float(row[2]), float(row[3]), row[4][5:])
    for row in data
]
flowers
```

```python
length, width = flowers[0].get_petal()
length, width
```
